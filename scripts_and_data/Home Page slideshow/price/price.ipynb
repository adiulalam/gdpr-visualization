```python
from bokeh.plotting import figure, output_file, show, save, ColumnDataSource
from bokeh.models.tools import HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Blues8
from bokeh.embed import components
import pandas
import pandas as pd
```

```python
df = pandas.read_csv('price.csv')

# Create ColumnDataSource from data frame
source = ColumnDataSource(df)

output_file('price.html')
```

```python
# Device list
Device_list = source.data['Device'].tolist()


# Add plot
p = figure(
    y_range=Device_list,
    plot_width=950,
    plot_height=500,
    title='Price Per Smart Device',
    x_axis_label='Price (Â£)',
    tools="pan,box_select,zoom_in,zoom_out,save,reset"
)

# Render glyph
p.hbar(
    y='Device',
    right='Price',
    left=0,
    height=0.4,
    fill_color=factor_cmap(
      'Device',
      palette=Blues8,
      factors=Device_list
    ),
    fill_alpha=0.9,
    source=source,
    legend_label='Smart Devices'
)

# Add Legend
p.legend.orientation = 'vertical'
p.legend.location = 'top_right'
p.legend.label_text_font_size = '10px'

# Add Tooltips
hover = HoverTool()
hover.tooltips = """
  <div>
    <h3>@Device</h3>
    <div><strong>Brand: </strong>@Brand</div>
    <div><strong>Price: Â£</strong>@Price</div>
    <div><img src="@Image" alt="" width="100" /></div>
  </div>
"""
p.add_tools(hover)

# Show results
show(p)

# Save file
save(p)

# Print out div and script
script, div = components(p)
print(div)
print(script)
```
